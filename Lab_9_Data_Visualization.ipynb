**Data visualization** is the process of graphically representing data.

It is the act of translating data into a visual context, which can be done using charts, plots, animations, infographics, etc.

The idea behind it is to make it easier for us to identify trends, outliers, and patterns in data.

**What is Matplotlib?**
Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

Matplotlib was created by John D. Hunter.

Matplotlib is open source and we can use it freely.

Matplotlib is mostly written in python, a few segments are written in C, Objective-C and Javascript for Platform compatibility.

**Import matplotlib and its sub module pyplot**

```python
  import matplotlib.pyplot as plt
  import numpy as np
  import pandas as pd
```

**Plotting x and y points**

The plot() function is used to draw points (markers) in a diagram.

By default, the plot() function draws a line from point to point.

The function takes parameters for specifying points in the diagram.

Parameter 1 is an array containing the points on the x-axis.

Parameter 2 is an array containing the points on the y-axis.

```python
x = np.array([1, 8])
y = np.array([3, 10])

plt.plot(x, y)
plt.show()
```

**Add labels and title (optional but recommended):**

 Enhance readability by adding labels to the x and y axes and a title to the plot.


    plt.xlabel("X-axis Label")
    plt.ylabel("Y-axis Label")
    plt.title("My Line Chart")

Display the plot: Use **plt.show()** to render the chart.

    plt.show()

***Matplotlib offers extensive customization for line charts, including:***

Line style: linestyle='--' (dashed), '-.' (dash-dot), ':' (dotted)

Line color: color='red', 'blue', '#FF0000' (hex code)

Line width: linewidth=2

Markers: marker='o' (circle), '^' (triangle), 's' (square)

Adding a legend: plt.legend(['My Data Series'])

Adding grid lines: plt.grid(True)

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.figure(figsize=(8,5))
plt.plot(x,y,color='red',linestyle='-.',marker='^',linewidth=2)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("line chart")
plt.legend(['my data series'])
plt.grid(False)
plt.show()
```

**Bar Chart**

**A bar chart in matplotlib (a popular Python plotting library) is used to display categorical data with rectangular bars representing the value of each category.**




```python
categories=['A','B','C','D','E']
values=[5,2,3,4,6]
plt.figure(figsize=(8,5))
plt.bar(categories,values,color='green')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('Bar chart')
plt.show()
```

Histogram

Histograms are used to represent the distribution of a dataset.

It divides data into bins and count the number of data points in each bin.

```python
# Sample data
#data = [12, 15, 13, 17, 16, 16, 19, 12, 14, 16, 14, 17, 18, 18, 15, 14]
data=[2, 3, 5, 5, 7, 8, 8, 8, 9, 10]
# Create histogram
plt.hist(data, bins=16, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Simple Histogram Example')
plt.show()
```

**Scatter plot**

A scatter plot in matplotlib is used to visualize the relationship between two numerical variables, with each point representing a pair of values (x and y).

plt.scatter() creates the scatter plot.

x and y are lists or arrays of the two variables.

plt.xlabel(), plt.ylabel(), and plt.title() set the axis labels and plot title.

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [5, 7, 4, 6, 8]

# Create scatter plot
plt.scatter(x, y)
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.title('Simple Scatter Plot Example')
plt.show()
```

**Pie chart**

A pie chart in matplotlib displays the proportion of each category as slices of a circle, making it ideal for visualizing part-to-whole relationships.

plt.pie() draws the pie chart.

labels names each slice.

sizes defines the size of each slice.

autopct displays the percentage on each slice.

startangle rotates the starting position.

plt.axis('equal') keeps the chart circular.

plt.title() sets the chart title.

Use a pie chart when you want to emphasize how each category contributes to the whole.

```python
# Sample data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors=['red','blue','yellow','pink']

# Create pie chart
plt.pie(sizes, colors=colors, labels=labels, autopct='%1.5f%%', startangle=140)
plt.title('Simple Pie Chart Example')
plt.axis('equal')  # Ensures the pie is drawn as a circle
plt.show()
```

**Box plot**

A box plot in matplotlib is a graphical tool used to display the distribution, spread, and outliers of a dataset.

It shows the median, quartiles, and possible outliers of your data.

plt.boxplot(data) creates the box plot.

The box shows the interquartile range (middle 50% of the data).

The line inside the box marks the median.

Whiskers show the extent of the rest of the data, and points outside are outliers.

plt.title() and plt.ylabel() provide chart labeling.

Box plots are great for comparing distributions and spotting variability or unusual values in your data.

```python
# Sample data
#data = [7, 8, 5, 6, 7, 9, 12, 5, 7, 8, 6]
data=[6, 7, 8, 8, 9, 10, 12]
# Create box plot
plt.boxplot(data)
plt.title('Simple Box Plot Example')
plt.ylabel('Values')
plt.show()
```

**Heatmap**

A heatmap in matplotlib is used to visualize data in matrix form, where values are represented as colors.

This makes it easy to spot patterns, correlations, or anomalies in datasets, especially for correlation matrices or frequency tables.

plt.imshow() displays the matrix as an image.

cmap='viridis' chooses the color map.

plt.colorbar() adds a scale showing the mapping from color to value.

plt.title() sets the chart title.

This visual approach helps you quickly see high and low values and overall structure in your data.

```python
# Sample data: 10x10 matrix
data = np.random.rand(10, 10)

# Create heatmap
plt.imshow(data, cmap='viridis', interpolation='nearest')
plt.colorbar()  # Show color scale
plt.title('Simple Heatmap Example')
plt.show()
```
